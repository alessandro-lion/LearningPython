# Primo ciclo, costruzione where per nomi solo femminili

```python
#PRIMO CICLO
#stampa solo nomi femminili
print("UPDATE YOURTABLE SET SESSO ='F' WHERE NOME IN ('EVA' ");
for x in nomiverificabili:
  
  for url in search('Nome ' +x, stop=6):
     if 'wikipedia' in url:
       req = requests.get(url).text;
       gotone=0
       for ss in sf:
         if ss in req:
           gotone=1
           break
       if not ('maschile e femminile' in req):
         if gotone==1:    
           print(", '"+x+"'")
           nomipotenzialmentemaschili.remove(x)
           #print(req)  
           break
print(") ")
```

# Secondo ciclo, costruzione where per nomi sicuramente maschili

```python
#test stampa solo nomi maschili
print("UPDATE YOURTABLE SET SESSO ='M' WHERE NOME IN ('ADAMO' ");
for x in nomiverificabili:
  for url in search('Nome ' +x, stop=6):
     if 'wikipedia' in url:
       req = requests.get(url).text
      
       if tt in req:
         if not ('Femminili: '+x.capitalize() in req):
           if not ('maschile e femminile' in req):
             print(", '"+x+"'");
             #print(req)  
             break
print(") ")
```
